```python
••••••••import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings(action='ignore')
```

```python
data = pd.read_csv('CAERS_ASCII_2004_2017Q2.csv')
data
```

```python
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

```python
newborns_products = data.loc[data['CI_Age Unit'] == 'Month(s)']
newborns_products = newborns_products.drop(newborns_products[newborns_products.CI_Gender == "Not Available"].index)
newborns_products.reset_index(drop = True, inplace = True)
```

```python
uns = pd.Series(newborns_products['PRI_FDA Industry Name'].unique())
uns
```

```python
new = newborns_products.groupby('PRI_FDA Industry Name').count()
new = new.rename(columns={"RA_Report #": "Cases_Reported", "PRI_FDA Industry Name": "Industry Name"})
new = new.filter(['PRI_FDA Industry Name', 'Cases_Reported'])
new.sort_values(by=['Cases_Reported'], inplace=True)
new
```
