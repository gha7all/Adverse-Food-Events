```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings(action='ignore')
```

```python
data = pd.read_csv('CAERS_ASCII_2004_2017Q2.csv')
data
```

```python
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

```python
newborns_products = data.loc[data['CI_Age Unit'] == 'Month(s)']
newborns_products = newborns_products.drop(newborns_products[newborns_products.CI_Gender == "Not Available"].index)
newborns_products.reset_index(drop = True, inplace = True)
newborns_products
```
